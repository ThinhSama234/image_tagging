# Parse label từ ADE20K JSON

```python
def get_labels_from_json(json_path):
    with open(json_path, "r") as f:
        data = json.load(f)

    labels = set()

    # object labels
    for obj in data.get("objects", []):
        labels.add(obj["classTitle"].lower())

    # scene tags (optional)
    for tag in data.get("tags", []):
        if tag["name"] == "scene":
            for s in tag["value"].split(","):
                labels.add(s.strip().lower())

    return list(labels)
```

```python
def normalize_label(label):
    return label.lower().strip()

def normalize_list(lst):
    return list(set([normalize_label(x) for x in lst]))
```

```python
def parse_caption_to_tags(caption):
    try:
        graph = sng_parser.parse(caption)
    except:
        return []

    tags = []

    # entities
    for ent in graph["entities"]:
        tags.append(ent["lemma_head"])

    # relations (verbs / prepositions)
    for rel in graph["relations"]:
        tags.append(rel["lemma_relation"])

    return tags
```

```python
def normalize_tags(tags):
    clean_tags = []

    for t in tags:
        t = t.lower().strip()

        # bỏ stopword
        if t in stop_words:
            continue

        clean_tags.append(t)

    return clean_tags
```

```python
import pandas as pd
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# =========================
# Utils
# =========================
def normalize_tag(tag):
    tag = tag.strip().lower()
    tag = tag.replace("_", " ")
    tag = " ".join(tag.split())
    return tag

def lemma_tag(tag):
    return " ".join([lemmatizer.lemmatize(w) for w in tag.split()])

def parse_tags(tag_str):
    return [t.strip() for t in tag_str.split("|")]

# =========================
# Soft match
# =========================
def soft_match(p, gt_set):
    for g in gt_set:
        # exact
        if p == g:
            return True
        # substring
        if p in g or g in p:
            return True
    return False

# =========================
# Load data
# =========================
df = pd.read_csv("/kaggle/working/flickr30k.csv")

# =========================
# Compute P/R (soft match)
# =========================
tp, fp, fn = 0, 0, 0

for _, row in df.iterrows():
    gt_tags = parse_tags(row["gt"])
    pred_tags = parse_tags(row["pred"])

    gt_set = set([lemma_tag(normalize_tag(t)) for t in gt_tags])
    pred_set = set([lemma_tag(normalize_tag(t)) for t in pred_tags])

    matched_gt = set()

    # ---- TP / FP ----
    for p in pred_set:
        found = False
        for g in gt_set:
            if p == g or p in g or g in p:
                tp += 1
                matched_gt.add(g)
                found = True
                break

        if not found:
            fp += 1

    # ---- FN ----
    fn += len(gt_set - matched_gt)

# =========================
# Final metrics
# =========================
precision = tp / (tp + fp + 1e-8)
recall = tp / (tp + fn + 1e-8)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
```

# mapping wordnet + exact match

```python
import pandas as pd
from nltk.corpus import wordnet as wn
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# =========================
# Utils
# =========================
def normalize_tag(tag):
    tag = tag.strip().lower()
    tag = tag.replace("_", " ")
    tag = " ".join(tag.split())
    return tag

def lemma_tag(tag):
    return " ".join([lemmatizer.lemmatize(w) for w in tag.split()])

def parse_tags(tag_str):
    return [t.strip() for t in tag_str.split("|")]

# =========================
# WordNet mapping
# =========================
def get_synonyms(tag):
    syns = set()
    for syn in wn.synsets(tag):
        for lemma in syn.lemmas():
            syns.add(lemma.name().replace("_", " "))
    return syns

def map_to_wordnet(tags):
    """
    Map mỗi tag → tập synonym
    """
    mapped = set()

    for t in tags:
        t = normalize_tag(t)
        t = lemma_tag(t)

        syns = get_synonyms(t)

        if len(syns) > 0:
            mapped.update(syns)
        else:
            mapped.add(t)

    return mapped

# =========================
# Load data
# =========================
df = pd.read_csv("/kaggle/working/flickr30k.csv")

# =========================
# Compute P/R (WordNet exact)
# =========================
tp, fp, fn = 0, 0, 0

for _, row in df.iterrows():
    gt_tags = parse_tags(row["gt"])
    pred_tags = parse_tags(row["pred"])

    gt_set = map_to_wordnet(gt_tags)
    pred_set = map_to_wordnet(pred_tags)

    # exact match trên synonym space
    inter = pred_set & gt_set

    tp += len(inter)
    fp += len(pred_set - gt_set)
    fn += len(gt_set - pred_set)

# =========================
# Final metrics
# =========================
precision = tp / (tp + fp + 1e-8)
recall = tp / (tp + fn + 1e-8)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
```

```python
import pandas as pd
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')

# đọc CSV
df = pd.read_csv("/kaggle/input/datasets/nguyenbon/ade-20k/ade-20k.csv")

# khởi tạo lemmatizer
lemmatizer = WordNetLemmatizer()

# hàm chuẩn hóa và lemmatize label
def normalize_label(label):
    if not label:
        return ""
    label = label.lower()                 # lowercase
    label = " ".join(label.split())       # chuẩn hóa khoảng trắng
    label = lemmatizer.lemmatize(label)   # lemmatize
    return label

# hàm tách và chuẩn hóa tag
def split_and_normalize(tag_str):
    if pd.isna(tag_str):
        return []
    tags = [normalize_label(t) for t in tag_str.split("|")]
    return tags

# tách tag và chuẩn hóa cho gt và pred
df["gt_tags"] = df["gt"].apply(split_and_normalize)
df["pred_tags"] = df["pred"].apply(split_and_normalize)

# tính Precision/Recall mỗi ảnh
def precision_recall(gt_list, pred_list):
    gt_set = set(gt_list)
    pred_set = set(pred_list)
    if not pred_set:
        precision = 0.0
    else:
        precision = len(gt_set & pred_set) / len(pred_set)
    if not gt_set:
        recall = 0.0
    else:
        recall = len(gt_set & pred_set) / len(gt_set)
    return precision, recall

df["precision"], df["recall"] = zip(*df.apply(lambda row: precision_recall(row["gt_tags"], row["pred_tags"]), axis=1))

# in kết quả trung bình
print("Precision trung bình:", df["precision"].mean())
print("Recall trung bình:", df["recall"].mean())
```
